```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
# ignore warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("data.csv")
```

```python
data.drop(["id", "Unnamed: 32"],axis=1,inplace=True)
data.head()
```

```python
M = data[data.diagnosis == "M"]
B = data[data.diagnosis == "B"]
```

```python
plt.scatter(M.radius_mean, M.texture_mean, color = "red", label="bad", alpha = 0.3)
plt.scatter(B.radius_mean, B.texture_mean, color = "green", label="good", alpha = 0.3)
plt.xlabel("radius_mean")
plt.ylabel("texture_mean")
plt.legend()
plt.show()
```

```python
data.diagnosis = [1 if each == "M" else 0 for each in data.diagnosis]
y = data.diagnosis.values
x_data = data.drop(["diagnosis"],axis=1)
```

```python
# normalization
x = (x_data - np.min(x_data))/(np.max(x_data) - np.min(x_data))
```

```python
# train - test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3, random_state = 1)
```

```python
# knn model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 6) # n_neighbors = k
knn.fit(x_train, y_train)
prediction = knn.predict(x_test)
```

```python
print(" {} nn score: {}".format(6, knn.score(x_test, y_test)))
```

```python
# find k value
score_list = []
for each in range(1,15):
    knn2 = KNeighborsClassifier(n_neighbors = each)
    knn2.fit(x_train,y_train)
    score_list.append(knn2.score(x_test,y_test))
    
plt.plot(range(1,15),score_list)
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.show()
```

# Conclusion
* K value gives better results if k value is between 6 - 10
